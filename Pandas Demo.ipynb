# Filtering

```python
df.loc[filt]
```

```python
filt = (df['Country'] == 'Pakistan') & (df['WebframeHaveWorkedWith'] == 'Django')

df.loc[filt]
```

```python
filt = (df['Country'] == 'Pakistan') | (df['WebframeHaveWorkedWith'] == 'Django')

df.loc[filt]
```

```python
# Adding a '~' sign before the filter would negate the filter. i.e

filt = (df['Country'] == 'Pakistan') | (df['WebframeHaveWorkedWith'] == 'Django')

df.loc[~filt, 'Country']
```

# Grouping and Aggregating
#### Splitting, Applying Functions & Combing Results

```python
df
```

```python
df['CompTotal'].median()
```

```python
df.median()
```

```python
df.describe()
```

```python
df['Country'].value_counts()
```

```python
country_group = df.groupby(['Country'])
```

```python
country_group['OpSys'].value_counts().loc['India']
```

```python
country_group['CompTotal'].agg(['mean','median']).loc['Pakistan']
```

```python
filt = df['Country'] == 'Pakistan'

df.loc[filt]['LanguageHaveWorkedWith'].str.contains('C').sum()
```

```python
country_group['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
country_group['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).head(10)
```

```python
country_respondent = df['Country'].value_counts()
country_respondent
```

```python
country_uses_python = country_group['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_uses_python
```

```python
python_df = pd.concat([country_respondent,country_uses_python], axis = 'columns', sort = False)
```

```python
python_df
```

```python
python_df.rename(columns = {'Country':'NumRespondents','LanguageHaveWorkedWith':'NumKnowsPython'}, inplace = True)
```

```python
python_df
```

```python
python_df['%KnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df
```

```python
python_df.sort_values(by = '%KnowsPython', ascending = False, inplace = True)
```
