### Rows

```python
# iloc ---> Rows by integer location
df.iloc[0] # 1st row (Series are simply defined as Column of Rows)
```

```python
df.iloc[[0,1]] # for accesing more than one row.
```

```python
df.iloc[[1,2], 6] # 1,2 are the number of rows we want to access and the number '6' is the cloumn number we want to access.
```

```python
#loc ----> searching by lables


df.loc[[1,2],'Employment']  # Here instead of passing the column number, we can simply pass the label of Column we want to access.
```

```python
# We can also access row data of multiple columns (i.e labels) by passing in [] such as

df.iloc[[1,2],[3,4]]

```

```python
# Also with the loc

df.loc[[1,2],['Country','US_State']]
```

```python
df.loc[1:2,'MainBranch':'Employment']  # Slicing
```
